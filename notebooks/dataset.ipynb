### Visualize some samples

```python
def display_sample(
    image : np.ndarray,
    masks : np.ndarray,
    label : str
) -> None:
    
    cols = 1 + masks.shape[-1]
    fig, ax = plt.subplots(1, cols, figsize=(14, 3))

    ax[0].imshow(image)
    ax[0].set_title('Image')
    ax[0].axis('off')

    for i in range(masks.shape[-1]):
        ax[i + 1].imshow(masks[:,:,i])
        ax[i + 1].set_title(f'Mask {i}')
        ax[i + 1].axis('off')

    fig.suptitle(label)
```

```python
samples_indices = np.random.randint(0, len(dataset), 5)

for i in samples_indices:
    sample = dataset[i]
    display_sample(sample['image'], sample['mask'], f"Id : {i}, Label : {sample['type']}")
```
